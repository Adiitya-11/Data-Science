```python
import matplotlib.pyplot as plt
%matplotlib inline

blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins =[80,100,125,150], rwidth = 0.8,color= 'green')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist([blood_sugar_men,blood_sugar_women] , bins =[80,100,125,150], rwidth = 0.8,color= ['green', 'red'], label=['men','women'])
plt.legend()
```

```python
plt.hist([blood_sugar_men,blood_sugar_women] , bins =[80,100,125,150], rwidth = 0.8,color= ['green', 'red'], label=['men','women'], orientation = 'horizontal')
plt.legend()
```
